```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex
```

```python
docs=SimpleDirectoryReader("./data/paul_graham/").load_data()
docs
```

### Defining model and embeddings

```python
from llama_index.llms.openai import OpenAI
from llama_index.embeddings.openai import OpenAIEmbedding
from llama_index.core import Settings
```

```python
Settings.llm = OpenAI(model="gpt-3.5-turbo", temperature=0.1)
Settings.embed_model = OpenAIEmbedding()
Settings.chunk_size = 512

# Behind the scenes, LlamaIndex automatically:
# 1. Looks for os.environ["OPENAI_API_KEY"]
# 2. Uses that key for authentication
# 3. No need to explicitly pass it
```

### Define VectorStore & QueryEngine

```python
index = VectorStoreIndex.from_documents(docs)
query_engine=index.as_query_engine()
```

```python
#  see your vector store details
print("Vector store type:", type(index._vector_store))
print("Storage context:", index._storage_context)
print("Number of nodes:", len(index.docstore.docs))
```

### Response

```python
res=query_engine.query("what does author say about html?")
```
